```python
import pandas as pd
```

```python
sns_1 = pd.read_csv('../data/sns2018_1.csv')
sns_2 = pd.read_csv('../data/sns2018_2.csv')
sns_3 = pd.read_csv('../data/sns2019_1.csv')
sns_4 = pd.read_csv('../data/sns2019_2.csv')
```

```python
sns_1.columns = ['id', 'date', 'bigcat', 'smallcat', 'sns']
sns_2.columns = ['id', 'date', 'bigcat', 'smallcat', 'sns']
sns_3.columns = ['id', 'date', 'bigcat', 'smallcat', 'sns']
sns_4.columns = ['id', 'date', 'bigcat', 'smallcat', 'sns']
```

```python
sns_2018 = pd.concat([sns_1, sns_2], axis=0)
sns_2019 = pd.concat([sns_3, sns_4], axis=0)
sale_all = pd.concat([sns_2018, sns_2019], axis=0)
```

```python
def pivot_sns(df):
    pivot = pd.pivot_table(data=df, index='date',columns='smallcat',values = 'sns',aggfunc='sum')
    pivot = pivot.reset_index()
    pivot.columns = [col +'_s' for col in pivot.columns]
    return pivot
```

```python
sns_pivot = pivot_sns(sale_all)
```

```python
def pivot_like(df, k):
    pivot = pd.pivot_table(data=df, index='date',columns='smallcat',values = k, aggfunc='sum')
    pivot = pivot.reset_index()
    pivot.columns = [col +'_b' for col in pivot.columns]
    return pivot
```

```python
def pivot_sns(df):
    pivot = pd.pivot_table(data=df, index='date',columns='smallcat',values = 'sns',aggfunc='sum')
    pivot = pivot.reset_index()
    pivot.columns = [col +'_s' for col in pivot.columns]
    return pivot
```

```python
# data = []
all_df = pd.DataFrame(data)

column = ["20대 남성 선호도 점수", "20대 여성 선호도 점수", 
          "30대 남성 선호도 점수", "30대 여성 선호도 점수",
          "40대 남성 선호도 점수", "40대 여성 선호도 점수", 
          "50대 남성 선호도 점수", "50대 여성 선호도 점수",
          "60대 남성 선호도 점수", "60대 여성 선호도 점수", ]

for k in column:
    like = data_all[["날짜", "대분류", "소분류", k]]
    sns_pivot = pivot_sns(sale_all)
    like.columns = ['date', 'bigcat', 'smallcat', k]
    like_piv= pivot_like(like, k)
    like_piv.fillna(0)
    total = pd.concat([like_piv, sns_pivot.iloc[:,1:]], axis=1)
    total_corr = total.corr()
    
    dic = {}

    for obj in total_corr.index[1:]:
        val = total_corr.loc[obj, obj[:-2]+'_s']
        if obj[:-2] not in dic.keys():
            dic[obj[:-2]]=val

    for obj, val in dic.items():
        if val >= 0.1:
            comment_info = [k, obj, val]
            all_df = all_df.append(pd.Series(comment_info), ignore_index=True)   
```
