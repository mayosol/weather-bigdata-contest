```python
import pandas as pd

X_train_knn = pd.read_csv("../data/X_train_weather_columns_6.csv", encoding='CP949')
X_test_knn = pd.read_csv("../data/X_test_weather_columns_6.csv", encoding='CP949')
y_train_knn = pd.read_csv("../data/y_train_weather_columns_6.csv", encoding='CP949')
y_test_knn = pd.read_csv("../data/y_test_weather_columns_6.csv", encoding='CP949')
```

```python
from sklearn.neighbors import KNeighborsRegressor

X_train_knn_6 = X_train_knn.drop(['날짜'], axis=1)
X_test_knn_6 = X_test_knn.drop(['날짜'], axis=1)
y_train_knn_total_sales = y_train_knn[['합계']]
y_train_knn_20m = y_train_knn[['20대 남성 선호도 점수']]
y_train_knn_20f = y_train_knn[['20대 여성 선호도 점수']]
y_train_knn_30m = y_train_knn[['30대 남성 선호도 점수']]
y_train_knn_30f = y_train_knn[['30대 여성 선호도 점수']]
y_train_knn_40m = y_train_knn[['40대 남성 선호도 점수']]
y_train_knn_40f = y_train_knn[['40대 여성 선호도 점수']]
y_train_knn_50m = y_train_knn[['50대 남성 선호도 점수']]
y_train_knn_50f = y_train_knn[['50대 여성 선호도 점수']]
y_train_knn_60m = y_train_knn[['60대 남성 선호도 점수']]
y_train_knn_60f = y_train_knn[['60대 여성 선호도 점수']]
knn_total_sales=KNeighborsRegressor()
knn_20m=KNeighborsRegressor()
knn_20f=KNeighborsRegressor()
knn_30m=KNeighborsRegressor()
knn_30f=KNeighborsRegressor()
knn_40m=KNeighborsRegressor()
knn_40f=KNeighborsRegressor()
knn_50m=KNeighborsRegressor()
knn_50f=KNeighborsRegressor()
knn_60m=KNeighborsRegressor()
knn_60f=KNeighborsRegressor()
knn_total_sales.fit(X_train_knn_6,y_train_knn_total_sales)
knn_total_sales_result=knn_total_sales.predict(X_test_knn_6)
knn_20m.fit(X_train_knn_6,y_train_knn_20m)
knn_20m_result=knn_20m.predict(X_test_knn_6)
knn_20f.fit(X_train_knn_6,y_train_knn_20f)
knn_20f_result=knn_20f.predict(X_test_knn_6)
knn_30m.fit(X_train_knn_6,y_train_knn_30m)
knn_30m_result=knn_30m.predict(X_test_knn_6)
knn_30f.fit(X_train_knn_6,y_train_knn_30f)
knn_30f_result=knn_30f.predict(X_test_knn_6)
knn_40m.fit(X_train_knn_6,y_train_knn_40m)
knn_40m_result=knn_40m.predict(X_test_knn_6)
knn_40f.fit(X_train_knn_6,y_train_knn_40f)
knn_40f_result=knn_40f.predict(X_test_knn_6)
knn_50m.fit(X_train_knn_6,y_train_knn_50m)
knn_50m_result=knn_50m.predict(X_test_knn_6)
knn_50f.fit(X_train_knn_6,y_train_knn_50f)
knn_50f_result=knn_50f.predict(X_test_knn_6)
knn_60m.fit(X_train_knn_6,y_train_knn_60m)
knn_60m_result=knn_60m.predict(X_test_knn_6)
knn_60f.fit(X_train_knn_6,y_train_knn_60f)
knn_60f_result=knn_60f.predict(X_test_knn_6)
```

```python
import joblib
import pandas as pd
joblib.dump(knn_20m, '../data/knn_20m.pkl')
joblib.dump(knn_20f, '../data/knn_20f.pkl')
joblib.dump(knn_30m, '../data/knn_30m.pkl')
joblib.dump(knn_30f, '../data/knn_30f.pkl')
joblib.dump(knn_40m, '../data/knn_40m.pkl')
joblib.dump(knn_40f, '../data/knn_40f.pkl')
joblib.dump(knn_50m, '../data/knn_50m.pkl')
joblib.dump(knn_50f, '../data/knn_50f.pkl')
joblib.dump(knn_60m, '../data/knn_60m.pkl')
joblib.dump(knn_60f, '../data/knn_60f.pkl')
joblib.dump(knn_total_sales, '../data/knn_total_sales.pkl')
pd.DataFrame(knn_20m_result).rename({0:'20대 남성 선호도 점수'}, axis='columns').to_csv('../data/knn_20m.csv')
pd.DataFrame(knn_20f_result).rename({0:'20대 여성 선호도 점수'}, axis='columns').to_csv('../data/knn_20f.csv')
pd.DataFrame(knn_30m_result).rename({0:'30대 남성 선호도 점수'}, axis='columns').to_csv('../data/knn_30m.csv')
pd.DataFrame(knn_30f_result).rename({0:'30대 여성 선호도 점수'}, axis='columns').to_csv('../data/knn_30f.csv')
pd.DataFrame(knn_40m_result).rename({0:'40대 남성 선호도 점수'}, axis='columns').to_csv('../data/knn_40m.csv')
pd.DataFrame(knn_40f_result).rename({0:'40대 여성 선호도 점수'}, axis='columns').to_csv('../data/knn_40f.csv')
pd.DataFrame(knn_50m_result).rename({0:'50대 남성 선호도 점수'}, axis='columns').to_csv('../data/knn_50m.csv')
pd.DataFrame(knn_50f_result).rename({0:'50대 여성 선호도 점수'}, axis='columns').to_csv('../data/knn_50f.csv')
pd.DataFrame(knn_60m_result).rename({0:'60대 남성 선호도 점수'}, axis='columns').to_csv('../data/knn_60m.csv')
pd.DataFrame(knn_60f_result).rename({0:'60대 여성 선호도 점수'}, axis='columns').to_csv('../data/knn_60f.csv')
pd.DataFrame(knn_total_sales_result).rename({0:'합계'}, axis='columns').to_csv('../data/expect_knn_total_sales.csv')
```

```python
df_knn_20m = pd.DataFrame(knn_20m_result).rename({0:'20대 남성 선호도 점수'}, axis='columns')
df_knn_20f = pd.DataFrame(knn_20f_result).rename({0:'20대 여성 선호도 점수'}, axis='columns')
df_knn_30m = pd.DataFrame(knn_30m_result).rename({0:'30대 남성 선호도 점수'}, axis='columns')
df_knn_30f = pd.DataFrame(knn_30f_result).rename({0:'30대 여성 선호도 점수'}, axis='columns')
df_knn_40m = pd.DataFrame(knn_40m_result).rename({0:'40대 남성 선호도 점수'}, axis='columns')
df_knn_40f = pd.DataFrame(knn_40f_result).rename({0:'40대 여성 선호도 점수'}, axis='columns')
df_knn_50m = pd.DataFrame(knn_50m_result).rename({0:'50대 남성 선호도 점수'}, axis='columns')
df_knn_50f = pd.DataFrame(knn_50f_result).rename({0:'50대 여성 선호도 점수'}, axis='columns')
df_knn_60m = pd.DataFrame(knn_60m_result).rename({0:'60대 남성 선호도 점수'}, axis='columns')
df_knn_60f = pd.DataFrame(knn_60f_result).rename({0:'60대 여성 선호도 점수'}, axis='columns')
df_knn_total_sales = pd.DataFrame(knn_total_sales_result).rename({0:'합계'}, axis='columns')
expect_knn_score = pd.concat([df_knn_20m, df_knn_20f, df_knn_30m, df_knn_30f, df_knn_40m, df_knn_40f, df_knn_50m, df_knn_50f, df_knn_60m, df_knn_60f], axis=1)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np
def rmse(y,pred):
    return np.sqrt(mean_squared_error(y,pred))
```

```python
rmse_knn_score = rmse(real_knn_score,expect_knn_score)
```
