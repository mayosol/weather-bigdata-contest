```python
import pandas as pd

X_train_linear = pd.read_csv("../data/X_train_weather_columns_6.csv", encoding='CP949')
X_test_linear = pd.read_csv("../data/X_test_weather_columns_6.csv", encoding='CP949')
y_train_linear = pd.read_csv("../data/y_train_weather_columns_6.csv", encoding='CP949')
y_test_linear = pd.read_csv("../data/y_test_weather_columns_6.csv", encoding='CP949')
```

```python
from sklearn.linear_model import LinearRegression

X_train_linear_6 = X_train_linear.drop(['날짜'], axis=1)
X_test_linear_6 = X_test_linear.drop(['날짜'], axis=1)
y_train_linear_total_sales = y_train_linear[['합계']]
y_train_linear_20m = y_train_linear[['20대 남성 선호도 점수']]
y_train_linear_20f = y_train_linear[['20대 여성 선호도 점수']]
y_train_linear_30m = y_train_linear[['30대 남성 선호도 점수']]
y_train_linear_30f = y_train_linear[['30대 여성 선호도 점수']]
y_train_linear_40m = y_train_linear[['40대 남성 선호도 점수']]
y_train_linear_40f = y_train_linear[['40대 여성 선호도 점수']]
y_train_linear_50m = y_train_linear[['50대 남성 선호도 점수']]
y_train_linear_50f = y_train_linear[['50대 여성 선호도 점수']]
y_train_linear_60m = y_train_linear[['60대 남성 선호도 점수']]
y_train_linear_60f = y_train_linear[['60대 여성 선호도 점수']]
linear_total_sales=LinearRegression()
linear_20m=LinearRegression()
linear_20f=LinearRegression()
linear_30m=LinearRegression()
linear_30f=LinearRegression()
linear_40m=LinearRegression()
linear_40f=LinearRegression()
linear_50m=LinearRegression()
linear_50f=LinearRegression()
linear_60m=LinearRegression()
linear_60f=LinearRegression()
linear_total_sales.fit(X_train_linear_6,y_train_linear_total_sales)
linear_total_sales_result=linear_total_sales.predict(X_test_linear_6)
linear_20m.fit(X_train_linear_6,y_train_linear_20m)
linear_20m_result=linear_20m.predict(X_test_linear_6)
linear_20f.fit(X_train_linear_6,y_train_linear_20f)
linear_20f_result=linear_20f.predict(X_test_linear_6)
linear_30m.fit(X_train_linear_6,y_train_linear_30m)
linear_30m_result=linear_30m.predict(X_test_linear_6)
linear_30f.fit(X_train_linear_6,y_train_linear_30f)
linear_30f_result=linear_30f.predict(X_test_linear_6)
linear_40m.fit(X_train_linear_6,y_train_linear_40m)
linear_40m_result=linear_40m.predict(X_test_linear_6)
linear_40f.fit(X_train_linear_6,y_train_linear_40f)
linear_40f_result=linear_40f.predict(X_test_linear_6)
linear_50m.fit(X_train_linear_6,y_train_linear_50m)
linear_50m_result=linear_50m.predict(X_test_linear_6)
linear_50f.fit(X_train_linear_6,y_train_linear_50f)
linear_50f_result=linear_50f.predict(X_test_linear_6)
linear_60m.fit(X_train_linear_6,y_train_linear_60m)
linear_60m_result=linear_60m.predict(X_test_linear_6)
linear_60f.fit(X_train_linear_6,y_train_linear_60f)
linear_60f_result=linear_60f.predict(X_test_linear_6)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np
def rmse(y,pred):
    return np.sqrt(mean_squared_error(y,pred))
```
