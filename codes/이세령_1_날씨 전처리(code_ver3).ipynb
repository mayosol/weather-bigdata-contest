날씨 변수 : 요소는 평균기온, 최저기온, 최저기온 시각, 최고기온, 최고기온 시각, 강수 계속시간, 1시간 최다강수량, 1시간 최다 강수량 시각, 일강수량, 평균풍속, 풍정합, 최다풍향, 최소 상대습도, 최소 상대습도 시각, 평균 상대습도  
미세먼지 등 변수 : 초미세먼지, 미세먼지, 오존  
이 중 최저기온 시각, 최고기온 시각, 1시간 최다 강수량 시각, 최소 상대습도 시간은 1시간을 간격으로 원-핫 인코딩 처리를 했으며, 최다풍향은 16방위를 기준으로 원-핫 인코딩하였다.  
그 후 지역별 인구비율 * 날씨/미세먼지 변수들을 곱해 계산했다. 이 후 최종파일명은 data 폴더 내 weahter_final이다.

최다풍향 처리

```python
import pandas as pd
```

```python
# load
mwd = pd.read_csv('../data/mwd.csv',encoding='cp949')
```

```python
mwd.shape
```

```python
mwd_ohe = pd.get_dummies(mwd['최다풍향(16방위)'])
```

```python
# ohe 후 칼럼명 변경
mwd_ohe.columns = ['최다풍향북북동','최다풍향북동','최다풍향동북동','최다풍향동','최다풍향동남동','최다풍향남동','최다풍향남남동','최다풍향남','최다풍향남남서','최다풍향남서','최다풍향서남서','최다풍향서','최다풍향서북서','최다풍향북서','최다풍향북북서','최다풍향북']
```

```python
mwd_ohe.head()
```

```python
mwd_fin = pd.concat([mwd,mwd_ohe], axis=1)
```

```python
mwd_fin.head()
```

```python
mwd_fin.columns
```

```python
mwd_fin.drop(['최다풍향(16방위)','평균 상대습도(%)'],axis=1,inplace=True)
```

```python
mwd_fin.head()
```

```python
mwd_fin.to_csv('mwd_fin.csv', encoding='cp949')
```

시각 데이터 처리

```python
weather_time = pd.read_csv('../data/weather_time.csv',encoding='cp949')
```

```python
weather_time.head()
```

```python
weather_time.columns
```

```python
weather_time.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
weather_time.head()
```

```python
weather_ohe_1 = pd.get_dummies(weather_time['최저기온 시각(hhmi)'])
weather_ohe_2 = pd.get_dummies(weather_time['최고기온 시각(hhmi)'])
weather_ohe_3 = pd.get_dummies(weather_time['1시간 최다 강수량 시각(hhmi)'])
weather_ohe_4 = pd.get_dummies(weather_time['최소 상대습도 시각(hhmi)'])
```

```python
weather_ohe_1 = weather_time.join(weather_ohe_1.add_prefix('최저기온 시각(hh)_'))
weather_ohe_2 = weather_time.join(weather_ohe_2.add_prefix('최고기온 시각(hh)'))
weather_ohe_3 = weather_time.join(weather_ohe_3.add_prefix('1시간 최다 강수량 시각(hh)'))
weather_ohe_4 = weather_time.join(weather_ohe_4.add_prefix('최소 상대습도 시각(hh)'))
```

```python
weather_ohe_3.head()
```

```python
weather_ohe_1.columns
```

```python
weather_ohe_3.columns
```

```python
# 필요없는 칼럼 제거
weather_ohe_1.drop(['날짜','지역','최저기온 시각(hhmi)', '최고기온 시각(hhmi)', '강수 계속시간(hr)',
       '1시간 최다 강수량 시각(hhmi)', '최소 상대습도 시각(hhmi)'],axis = 1)
weather_ohe_2.drop(['날짜','지역','최저기온 시각(hhmi)', '최고기온 시각(hhmi)', '강수 계속시간(hr)',
       '1시간 최다 강수량 시각(hhmi)', '최소 상대습도 시각(hhmi)'],axis = 1)
weather_ohe_3.drop(['날짜','지역','최저기온 시각(hhmi)', '최고기온 시각(hhmi)', '강수 계속시간(hr)',
       '1시간 최다 강수량 시각(hhmi)', '최소 상대습도 시각(hhmi)'],axis = 1)
weather_ohe_4.drop(['날짜','지역','최저기온 시각(hhmi)', '최고기온 시각(hhmi)', '강수 계속시간(hr)',
       '1시간 최다 강수량 시각(hhmi)', '최소 상대습도 시각(hhmi)'],axis = 1)
```

```python
weather_time_fin = pd.concat([weather_ohe_1,weather_ohe_2,weather_ohe_3,weather_ohe_4],axis=1)
```

```python
weather_time_fin.shape
```

```python
weather_time_fin.columns
```

```python
# 필요없는 칼럼 제거
weather_time_fin.drop([ '최저기온 시각(hhmi)', '최고기온 시각(hhmi)', '강수 계속시간(hr)',
       '1시간 최다 강수량 시각(hhmi)', '최소 상대습도 시각(hhmi)'],axis=1,inplace=True)
```

```python
weather_time_fin.head()
```

```python
# csv로 내보내기
weather_time_fin.to_csv('weather_time_fin.csv',encoding='cp949')
```
