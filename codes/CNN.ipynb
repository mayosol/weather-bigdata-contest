```python
# 라이브러리 사용
import tensorflow as tf
import pandas as pd
import numpy as np
 
###########################
# 데이터를 준비합니다. 
X_train_cnn = pd.read_csv("../data/X_train_weather_columns_6.csv", encoding='CP949').drop('날짜', axis=1)
X_test_cnn = pd.read_csv("../data/X_test_weather_columns_6.csv", encoding='CP949').drop('날짜', axis=1)
y_train_cnn = pd.read_csv("../data/y_train_weather_columns_6.csv", encoding='CP949')[['20대 남성 선호도 점수','20대 여성 선호도 점수','30대 남성 선호도 점수','30대 여성 선호도 점수','40대 남성 선호도 점수','40대 여성 선호도 점수','50대 남성 선호도 점수','50대 여성 선호도 점수','60대 남성 선호도 점수','60대 여성 선호도 점수', '합계']]
y_test_cnn = pd.read_csv("../data/y_test_weather_columns_6.csv", encoding='CP949')[['20대 남성 선호도 점수','20대 여성 선호도 점수','30대 남성 선호도 점수','30대 여성 선호도 점수','40대 남성 선호도 점수','40대 여성 선호도 점수','50대 남성 선호도 점수','50대 여성 선호도 점수','60대 남성 선호도 점수','60대 여성 선호도 점수', '합계']]
X_train_lenet = np.array(X_train_cnn).reshape(56268, 15, 9)
X_test_lenet = np.array(X_test_cnn).reshape(24116, 15, 9)
y_train_lenet = np.array(y_train_cnn).reshape(56268,11)
print(X_train_lenet.shape, X_test_lenet.shape)
 
###########################
# 모델을 완성합니다. 
X = tf.keras.layers.Input(shape=[15, 9, 1])
 
H = tf.keras.layers.Conv2D(6, kernel_size=3, activation='swish')(X)
 
H = tf.keras.layers.Conv2D(16, kernel_size=3, activation='swish')(H)
 
H = tf.keras.layers.Flatten()(H)
H = tf.keras.layers.Dense(120, activation='swish')(H)
H = tf.keras.layers.Dense(84, activation='swish')(H)
Y = tf.keras.layers.Dense(11, activation='softmax')(H)
 
model = tf.keras.models.Model(X, Y)
model.compile(loss='MeanSquaredError', metrics='MeanSquaredError')
 
###########################
# 모델을 학습하고
model.fit(X_train_lenet, y_train_lenet, epochs=10)
 
###########################
# 모델을 이용합니다. 
pred = model.predict(X_test_lenet)
y_test_lenet_predict=pd.DataFrame(pred)

# 모델 확인
model.summary()
```
