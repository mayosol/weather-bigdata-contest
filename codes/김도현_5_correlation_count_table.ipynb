```python
import pandas as pd
```

```python
corr_df = pd.read_csv('../data/sale_weather_correlation.csv')
```

```python
#상관계수에 절댓값을 씌워서 start와 end 사이에 있는 값들 count
#상관계수가 1인 것은 없지만 end가 1일때는 포함시켜서 하나도 빠짐업이 count
def n_corr_in(corr_df, start, end):
    if end == 1.0:
        count_info = ((corr_df.iloc[:, 1:].abs() >= start) & (corr_df.iloc[:, 1:].abs() <= end)).sum(axis = 1)
    else:
        count_info = ((corr_df.iloc[:, 1:].abs() >= start) & (corr_df.iloc[:, 1:].abs() < end)).sum(axis = 1)
    return count_info
```

```python
#범위에 따른 상관계수 리스트를 저장
n_corr_in_list = []
n_corr_in_list.append(n_corr_in(corr_df, 0.0, 0.2))
n_corr_in_list.append(n_corr_in(corr_df, 0.2, 0.4))
n_corr_in_list.append(n_corr_in(corr_df, 0.4, 0.6))
n_corr_in_list.append(n_corr_in(corr_df, 0.6, 0.8))
n_corr_in_list.append(n_corr_in(corr_df, 0.8, 1.0))
```

```python
#컬럼명 생성
corr_count_category = ['상관계수 절댓값({} 이상, {} 미만)'.format(i/10, round(i/10 + 0.2, 2)) for i in range(0, 10, 2)]

#corr_df에 추가
for i in range(5):
    corr_df['{}'.format(corr_count_category[i])] = n_corr_in_list[i]
```
