```python
import pandas as pd

X_train_svm = pd.read_csv("../data/X_train_weather_columns_6.csv", encoding='CP949')
X_test_svm = pd.read_csv("../data/X_test_weather_columns_6.csv", encoding='CP949')
y_train_svm = pd.read_csv("../data/y_train_weather_columns_6.csv", encoding='CP949')
y_test_svm = pd.read_csv("../data/y_test_weather_columns_6.csv", encoding='CP949')
```

```python
from sklearn.svm import SVR

X_train_svm_6 = X_train_svm.drop(['날짜'], axis=1)
X_test_svm_6 = X_test_svm.drop(['날짜'], axis=1)
y_train_svm_total_sales = y_train_svm[['합계']]
y_train_svm_20m = y_train_svm[['20대 남성 선호도 점수']]
y_train_svm_20f = y_train_svm[['20대 여성 선호도 점수']]
y_train_svm_30m = y_train_svm[['30대 남성 선호도 점수']]
y_train_svm_30f = y_train_svm[['30대 여성 선호도 점수']]
y_train_svm_40m = y_train_svm[['40대 남성 선호도 점수']]
y_train_svm_40f = y_train_svm[['40대 여성 선호도 점수']]
y_train_svm_50m = y_train_svm[['50대 남성 선호도 점수']]
y_train_svm_50f = y_train_svm[['50대 여성 선호도 점수']]
y_train_svm_60m = y_train_svm[['60대 남성 선호도 점수']]
y_train_svm_60f = y_train_svm[['60대 여성 선호도 점수']]
svm_total_sales=SVR()
svm_20m=SVR()
svm_20f=SVR()
svm_30m=SVR()
svm_30f=SVR()
svm_40m=SVR()
svm_40f=SVR()
svm_50m=SVR()
svm_50f=SVR()
svm_60m=SVR()
svm_60f=SVR()
svm_total_sales.fit(X_train_svm_6,y_train_svm_total_sales)
svm_total_sales_result=svm_total_sales.predict(X_test_svm_6)
svm_20m.fit(X_train_svm_6,y_train_svm_20m)
svm_20m_result=svm_20m.predict(X_test_svm_6)
svm_20f.fit(X_train_svm_6,y_train_svm_20f)
svm_20f_result=svm_20f.predict(X_test_svm_6)
svm_30m.fit(X_train_svm_6,y_train_svm_30m)
svm_30m_result=svm_30m.predict(X_test_svm_6)
svm_30f.fit(X_train_svm_6,y_train_svm_30f)
svm_30f_result=svm_30f.predict(X_test_svm_6)
svm_40m.fit(X_train_svm_6,y_train_svm_40m)
svm_40m_result=svm_40m.predict(X_test_svm_6)
svm_40f.fit(X_train_svm_6,y_train_svm_40f)
svm_40f_result=svm_40f.predict(X_test_svm_6)
svm_50m.fit(X_train_svm_6,y_train_svm_50m)
svm_50m_result=svm_50m.predict(X_test_svm_6)
svm_50f.fit(X_train_svm_6,y_train_svm_50f)
svm_50f_result=svm_50f.predict(X_test_svm_6)
svm_60m.fit(X_train_svm_6,y_train_svm_60m)
svm_60m_result=svm_60m.predict(X_test_svm_6)
svm_60f.fit(X_train_svm_6,y_train_svm_60f)
svm_60f_result=svm_60f.predict(X_test_svm_6)
```

```python
import joblib
import pandas as pd
joblib.dump(svm_20m, '../data/svm_20m.pkl')
joblib.dump(svm_20f, '../data/svm_20f.pkl')
joblib.dump(svm_30m, '../data/svm_30m.pkl')
joblib.dump(svm_30f, '../data/svm_30f.pkl')
joblib.dump(svm_40m, '../data/svm_40m.pkl')
joblib.dump(svm_40f, '../data/svm_40f.pkl')
joblib.dump(svm_50m, '../data/svm_50m.pkl')
joblib.dump(svm_50f, '../data/svm_50f.pkl')
joblib.dump(svm_60m, '../data/svm_60m.pkl')
joblib.dump(svm_60f, '../data/svm_60f.pkl')
joblib.dump(svm_total_sales, '../data/svm_total_sales.pkl')
pd.DataFrame(svm_20m_result).rename({0:'20대 남성 선호도 점수'}, axis='columns').to_csv('../data/svm_20m.csv')
pd.DataFrame(svm_20f_result).rename({0:'20대 여성 선호도 점수'}, axis='columns').to_csv('../data/svm_20f.csv')
pd.DataFrame(svm_30m_result).rename({0:'30대 남성 선호도 점수'}, axis='columns').to_csv('../data/svm_30m.csv')
pd.DataFrame(svm_30f_result).rename({0:'30대 여성 선호도 점수'}, axis='columns').to_csv('../data/svm_30f.csv')
pd.DataFrame(svm_40m_result).rename({0:'40대 남성 선호도 점수'}, axis='columns').to_csv('../data/svm_40m.csv')
pd.DataFrame(svm_40f_result).rename({0:'40대 여성 선호도 점수'}, axis='columns').to_csv('../data/svm_40f.csv')
pd.DataFrame(svm_50m_result).rename({0:'50대 남성 선호도 점수'}, axis='columns').to_csv('../data/svm_50m.csv')
pd.DataFrame(svm_50f_result).rename({0:'50대 여성 선호도 점수'}, axis='columns').to_csv('../data/svm_50f.csv')
pd.DataFrame(svm_60m_result).rename({0:'60대 남성 선호도 점수'}, axis='columns').to_csv('../data/svm_60m.csv')
pd.DataFrame(svm_60f_result).rename({0:'60대 여성 선호도 점수'}, axis='columns').to_csv('../data/svm_60f.csv')
pd.DataFrame(svm_total_sales_result).rename({0:'합계'}, axis='columns').to_csv('../data/expect_svm_total_sales.csv')
```

```python
df_svm_20m = pd.DataFrame(svm_20m_result).rename({0:'20대 남성 선호도 점수'}, axis='columns')
df_svm_20f = pd.DataFrame(svm_20f_result).rename({0:'20대 여성 선호도 점수'}, axis='columns')
df_svm_30m = pd.DataFrame(svm_30m_result).rename({0:'30대 남성 선호도 점수'}, axis='columns')
df_svm_30f = pd.DataFrame(svm_30f_result).rename({0:'30대 여성 선호도 점수'}, axis='columns')
df_svm_40m = pd.DataFrame(svm_40m_result).rename({0:'40대 남성 선호도 점수'}, axis='columns')
df_svm_40f = pd.DataFrame(svm_40f_result).rename({0:'40대 여성 선호도 점수'}, axis='columns')
df_svm_50m = pd.DataFrame(svm_50m_result).rename({0:'50대 남성 선호도 점수'}, axis='columns')
df_svm_50f = pd.DataFrame(svm_50f_result).rename({0:'50대 여성 선호도 점수'}, axis='columns')
df_svm_60m = pd.DataFrame(svm_60m_result).rename({0:'60대 남성 선호도 점수'}, axis='columns')
df_svm_60f = pd.DataFrame(svm_60f_result).rename({0:'60대 여성 선호도 점수'}, axis='columns')
df_svm_total_sales = pd.DataFrame(svm_total_sales_result).rename({0:'합계'}, axis='columns')
expect_svm_score = pd.concat([df_svm_20m, df_svm_20f, df_svm_30m, df_svm_30f, df_svm_40m, df_svm_40f, df_svm_50m, df_svm_50f, df_svm_60m, df_svm_60f], axis=1)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np
def rmse(y,pred):
    return np.sqrt(mean_squared_error(y,pred))
```

```python
real_svm_score = y_test_svm[['20대 남성 선호도 점수','20대 여성 선호도 점수','30대 남성 선호도 점수','30대 여성 선호도 점수','40대 남성 선호도 점수','40대 여성 선호도 점수','50대 남성 선호도 점수','50대 여성 선호도 점수','60대 남성 선호도 점수','60대 여성 선호도 점수']]
```

```python
rmse_svm_score = rmse(real_svm_score,expect_svm_score)
```
